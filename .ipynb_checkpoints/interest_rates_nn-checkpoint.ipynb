```python
# import packages
%matplotlib inline
import warnings
import numpy as np
import pandas as pd
from numpy import random
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import Quandl
```

# Data Wrangling

```python
# read in 1-Month, 3-Month, 12-Month Treasury Bill data from Quandl
three_month = Quandl.get("FRED/DGS3MO")
six_month = Quandl.get("FRED/DGS6MO")
twelve_month = Quandl.get("FRED/DGS1")
```

```python
# reset indices 
three_month.reset_index(level=0, inplace=True)
six_month.reset_index(level=0, inplace=True)
twelve_month.reset_index(level=0, inplace=True)
```

```python
# limit 12-Month Treasuries to after 1982
twelve_month = twelve_month[(twelve_month['DATE'] >= '1982-01-04')]
twelve_month = twelve_month.reset_index(drop=True)
```

```python
# normalize data to heave mean = 0, std = 1
three_month['normal'] = (three_month['VALUE'] - three_month['VALUE'].mean(axis=0)) / three_month['VALUE'].std(axis=0, ddof=1)
six_month['normal'] = (six_month['VALUE'] - six_month['VALUE'].mean(axis=0)) / six_month['VALUE'].std(axis=0, ddof=1)
twelve_month['normal'] = (twelve_month['VALUE'] - twelve_month['VALUE'].mean(axis=0)) / twelve_month['VALUE'].std(axis=0, ddof=1)
```

```python
# merge into single DataFrame
data = three_month.merge(six_month, on = 'DATE')
data = data.merge(twelve_month, on = 'DATE')
```

```python
# rename columns
data.columns = ['date', '3_rate', '3_norm', '6_rate', '6_norm', '12_rate', '12_norm']
```

```python
# compute the difference between 3-Month and 6-Month rates
data['diff_3_6'] = abs(data['3_rate'] - data['6_rate'])
data['diff_3_6_norm'] = abs(data['3_norm'] - data['6_norm'])
```

## To start, let's look only at 3-Month and 6-Month rates

```python
# plot to see what the normalized data look like
figure, ax = plt.subplots(figsize=(16,10))
ax.plot(data['date'], data['3_norm'], label = '3-Month Rates', color = '#1FBEC3')
ax.plot(data['date'], data['6_norm'], label = '6-Month Rates', color = '#F57670')
ax.plot(data['date'], data['diff_3_6_norm'], label = 'Difference in Rates', color = '#ffc952')

ax.set_xlabel('Year')
ax.set_ylabel('Interest Rate')
ax.set_title('Normalized Interest Rates of US Treasury Bills (1982-2016)')

ax.legend()
```

```python
# make a scatter plot of rates
figure, ax = plt.subplots(figsize=(16,10))
ax.scatter(data['3_norm'], data['6_norm'],color = '#1FBEC3',  edgecolors='black')

ax.set_xlabel('3-Month Rates')
ax.set_ylabel('6-Month Rates')
ax.set_title('3-Month Rates vs 6-Month Rates')
```

### Just from looking at the scatter plot above, we can see that there is a very linear relationship between the 3-Month and 6-Month rates. For such a simple example, a NN is probably not necessary. Still, let's implement one anyway to see what happens.

# Neural Network
## Predicting 6-Month, using 3-Month

```python
# set random seed
np.random.seed(3102016)

# slice a third of dataset for training set
size = len(data)/3
train_rows = data.sample(size)

# assign training set and test set
train_data = data.ix[train_rows.index]
test_data = data.drop(train_rows.index)

print (len(train_data))
print (len(test_data))
```

```python
from function_approximator import FunctionApproximator

# define the function approximator
fa = FunctionApproximator(n_in=1, n_out=1, n_hidden=5)

# use normalized sample of 3-Month rates as x for training
x = (train_data['3_norm']).reshape((len(train_data), 1))

# use normalized sample of 6-Month rates as y for training
y = (train_data['6_norm']).reshape((len(train_data), 1))

# train the model
fa.train(x, y, learning_rate=0.05, n_epochs=200000, report_frequency=20000)
```

```python
# define x and y test vectors
x_test = test_data['3_norm'].reshape((len(test_data), 1))
y_test = test_data['6_norm'].reshape((len(test_data), 1))
```

```python
# print mean squared error of test set
y_bar = fa.get_y_pred(x_test)
print (((y_test - y_bar) ** 2).mean(axis=0))
```

```python
# use model on remaining 3-Month rates to predict remaining 6-Month rates
y_predict = fa.get_y_pred(x_test)
```

```python
# create lists for plotting
test_data_x = test_data['date'].tolist()
data_x = data['date'].tolist()
```

```python
# plot to see what the prediction looks like vs the actual data
figure, ax = plt.subplots(figsize=(16,10))
ax.plot(test_data_x, y_predict, label = 'NN Estimate of 6-Month Rates', color = '#1FBEC3')
ax.plot(data_x, data['6_norm'], label = '6-Month Rates', color = '#F57670')

ax.set_xlabel('Year')
ax.set_ylabel('Interest Rate')
ax.set_title('NN Estimate of 6-Month Rates + Actual 6-Month Rats')

ax.legend()
```

## Predicting difference between 6-Month and 3-Month, using 3-Month

```python
# define the function approximator
fa = FunctionApproximator(n_in=1, n_out=1, n_hidden=5)

# use normalized sample of 3-Month rates as x for training
x = (train_data['3_norm']).reshape((len(train_data), 1))

# use normalized sample of difference between 3-Month and 6-Month rates for training
y = (train_data['diff_3_6_norm']).reshape((len(train_data), 1))

# train the model
fa.train(x, y, learning_rate=0.05, n_epochs=200000, report_frequency=20000)
```

```python
# define x and y test vectors
x_test = test_data['3_norm'].reshape((len(test_data), 1))
y_test = test_data['diff_3_6_norm'].reshape((len(test_data), 1))
```

```python
# print mean squared error of test set
y_bar = fa.get_y_pred(x_test)
print (((y_test - y_bar) ** 2).mean(axis=0))
```

```python
# test model on remaining 3-Month rates to predict difference between 3-Month and 6-Month rates
y_predict = fa.get_y_pred(test_data['3_norm'].reshape((len(test_data), 1)))
```

```python
# create lists for plotting
test_data_x = test_data['date'].tolist()
data_x = data['date'].tolist()
```

```python
# plot to see what the prediction looks like vs the actual data
figure, ax = plt.subplots(figsize=(16,10))
ax.plot(data_x, data['diff_3_6_norm'], label = 'Difference in Rates', color = '#1FBEC3')
ax.plot(test_data_x, y_predict, label = 'NN Estimate of Difference in Rates', color = '#F57670', linewidth=6)

ax.set_ylim(-0.1, 0.6)
ax.set_xlabel('Year')
ax.set_ylabel('Interest Rate')
ax.set_title('NN Estimate of the Difference in Rates + Actual Difference in Rates')

ax.legend()
```

### The NN seems to be able to follow the general trend pretty well. However, it does not seem at all capable of predicting spikes in the difference between rates.

## Predicting 12-Month, using 3-Month and 6-Month

```python
# plot to see what the normalized data look like
figure, ax = plt.subplots(figsize=(16,10))
ax.plot(data['date'], data['3_norm'], label = '3-Month Rates', color = '#ffc952')
ax.plot(data['date'], data['6_norm'], label = '6-Month Rates', color = '#1FBEC3')
ax.plot(data['date'], data['12_norm'], label = '12-Month Rates', color = '#F57670')

ax.set_xlabel('Year')
ax.set_ylabel('Interest Rate')
ax.set_title('Normalized Interest Rates of US Treasury Bills (1982-2016)')

ax.legend()
```

```python
# slice train dataset and reshape into 2D array
x3_train = (train_data['3_norm']).reshape((len(train_data),1))
x6_train = (train_data['6_norm']).reshape((len(train_data),1))
x_train = np.concatenate((x3_train, x6_train), axis=1)

# slice test dataset and reshape into 2D array
x3_test = (test_data['3_norm']).reshape((len(test_data),1))
x6_test = (test_data['6_norm']).reshape((len(test_data),1))
x_test = np.concatenate((x3_test, x6_test), axis=1)
```

```python
# define the function approximator
fa = FunctionApproximator(n_in=2, n_out=1, n_hidden=5)

# use normalized sample of 12-Month rates as y for training
y = (train_data['12_norm']).reshape((len(train_data),1))

# train the model
fa.train(x_train, y, learning_rate=0.05, n_epochs=200000, report_frequency=20000)
```

```python
# define y test vector
y_test = test_data['12_norm'].reshape((len(test_data), 1))
```

```python
# print mean squared error of test set
y_bar = fa.get_y_pred(x_test)
print (((y_test - y_bar) ** 2).mean(axis=0))
```

```python
# test model on remaining 3-Month and 6-Month rates to predict 12-Month rates
y_predict = fa.get_y_pred(x_test)
```

```python
# create lists for plotting
test_data_x = test_data['date'].tolist()
data_x = data['date'].tolist()
```

```python
# plot to see what the prediction looks like vs the actual data
figure, ax = plt.subplots(figsize=(16,10))
ax.plot(test_data_x, y_predict, label = 'NN Estimate of 12-Month Rates', color = '#1FBEC3')
ax.plot(data_x, data['12_norm'], label = '12-Month Rates', color = '#F57670')

ax.set_xlabel('Year')
ax.set_ylabel('Interest Rate')
ax.set_title('NN Estimate of 12-Month Rates + Actual 12-Month Rats')

ax.legend()
```
